```python
from glob import glob

pos_test_img_list = glob('pos*') # including 120 negative images
neg_test_img_list = glob('neg*')
print("number of positive images: " + str(len(pos_test_img_list)))
print("number of negative images: " + str(len(neg_test_img_list)))
```

```python
import pandas as pd
df1 = pd.DataFrame()
df1['file_name'] = pos_test_img_list
df1['relevant'] = 1

df2 = pd.DataFrame()
df2['file_name'] = neg_test_img_list
df2['relevant'] = 0
df3 = pd.concat([df1,df2]).reset_index(drop=True)
df3
```

```python
%cd /home/studio-lab-user/DS-Projects/yolov7/yolov7/runs/detect/exp/labels
```

```python
def score_extraction(row):
    image_name = row["file_name"]
    # image_prefix = image_name.split('.')[0]
    image_prefix = image_name[:-4]
    image_prefix_txt = image_prefix + ".txt"
    try:
        data = pd.read_csv(image_prefix_txt, header = None, delimiter = ' ')
    except:
        return 0
    
    max_score = data[5].max()
    return max_score
    
df3["max_score"] = df3.apply(score_extraction, axis=1)
```

```python
df3.sort_values('max_score', ascending= False).head(50)
```

```python
from sklearn.metrics import roc_auc_score

def get_roc_auc_score(df3):
    roc_auc = roc_auc_score(df3['relevant'], df3['max_score'])
    
    return roc_auc

print("rocauc_score: ", get_roc_auc_score(df3))
```

```python
# Dataframe of true positives
tp = df3.loc[(df3['relevant'] == 1) & (df3['max_score'] >= 0.5)] 
print("True Positive: ", len(tp))
tn = df3.loc[(df3['relevant'] == 0) & (df3['max_score'] < 0.5)] 
print("True Negative: ", len(tn))
fp = df3.loc[(df3['relevant'] == 0) & (df3['max_score'] >= 0.5)] 
print("False Positive: ", len(fp))
fn = df3.loc[(df3['relevant'] == 1) & (df3['max_score'] < 0.5)] 
print("False Negative: ", len(fn))
precision = len(tp) / (len(tp) + len(fp))
print("Precision: ", precision)
recall = len(tp) / (len(tp) + len(fn))
print("Recall: ", recall)
```
